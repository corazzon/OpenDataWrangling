## 상관 관계

* 상관 계수 값: 상관 계수는 -1부터 1까지의 값을 가지며, 0에 가까울수록 두 변수 간의 선형 관계가 약하거나 없음을 의미합니다. 양수 값은 양의 상관 관계(한 변수가 증가하면 다른 변수도 증가)를 의미하고, 음수 값은 음의 상관 관계(한 변수가 증가하면 다른 변수는 감소)를 의미합니다.

* 강도: 상관 계수의 절대값이 클수록 두 변수 간의 선형 관계가 강력하다고 볼 수 있습니다. 일반적으로 0.7 이상의 상관 계수는 강력한 상관 관계를 나타냅니다.

* 방향: 상관 계수의 부호는 두 변수 간의 관계의 방향을 나타냅니다. 양의 상관 관계는 두 변수가 동시에 증가하거나 감소하는 경향이 있음을 의미하고, 음의 상관 관계는 한 변수가 증가하면 다른 변수는 감소하는 경향이 있음을 의미합니다.
